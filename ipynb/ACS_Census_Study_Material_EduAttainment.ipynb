Given percentage of (High school graduate or higher, Bachelor's degree or higher) return (<HS, =HS, BS+)

```python
def nominalize_edu(hsPlus,bsPlus):
    hsLess = 100-hsPlus 
    hsOnly = hsPlus - bsPlus
    return hsLess,hsOnly,bsPlus#,sum([hsLess,hsOnly,bsPlus])
```

```python
print nominalize_edu(87.3,29.8) # Female
print nominalize_edu(83.8,19.5) # Black 
print nominalize_edu(85.0,21.6) # Female and Black 
```

```python
def bsBinarize(hsPlus,bsPlus):
    hsLess = 100-hsPlus 
    hsOnly = hsPlus - bsPlus
    return hsLess+hsOnly,bsPlus#,sum([hsLess+hsOnly,bsPlus])
```

```python
print bsBinarize(87.3,29.8) # Female
print bsBinarize(83.8,19.5) # Black 
print bsBinarize(85.0,21.6) # Female and Black 
```

```python
def multibar(ratios,subpopulation_name,bar_names):
    fig, ax = plt.subplots()
    ind = np.arange(len(bar_names))
    width = 0.3
    barlist = ax.bar(ind, ratios, width)
    ax.set_xticks(ind)
    ax.set_xticklabels(bar_names,fontsize=13)#, rotation='vertical'
    ax.set_title(subpopulation_name)
    ax.set_ylabel("Percentage",fontsize=13)
    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 0.5*height,
                    '%.2f' % float(height),
                    ha='center', va='bottom',color="white",fontweight='bold',fontsize=13)

    autolabel(barlist)
    ax.set_ylim(0,100)
```

```python
vals = [86.7,29.8]
subpopulation_name= 'Overall'
multibar(vals,subpopulation_name,['>=HS', '>=B'])
multibar(nominalize_edu(*vals),subpopulation_name,['<HS', '=HS', 'B+'])
multibar(bsBinarize(*vals),subpopulation_name,["<Bachelor's", "$\geq$ Bachelor's"])
```
