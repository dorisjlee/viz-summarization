```python
import os 
```

```python
experiment_name = "vary_dataset_ip"
if (not (os.path.exists("dashboards/"+experiment_name))):
    os.mkdir("dashboards/"+experiment_name)
for fname in glob.glob("dashboards/json/{}/*".format(experiment_name)):
    datasetname, xAxis,algo,dist,ic,ip,k,nbar = fname[:-5].split("/")[-1].split("_")
    ic = float(ic[2:])
    ip = float(ip[2:])
    k = int(k[1:])
    nbar = int(nbar[4:])
    with open(fname, 'r') as myfile:
        nodeDicStr=myfile.read().replace('\n', '').decode('string_escape')
    generateDashboard(nodeDicStr,xAxis,"COUNT",nbar,"{} {} [k={}] ip={}".format(datasetname,algo,k,ip),experiment_name+"/"+fname[:-5].split("/")[-1]+".pdf")
```
